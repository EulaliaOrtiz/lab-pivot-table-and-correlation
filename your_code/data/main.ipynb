#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here

employee.groupby(['Department']).agg({'Salary':np.mean}).reset_index(). rename(columns={'Salary':'Average salary'})
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2
# your answer here
# The syntax is 
# in the index argument you put the columns you want to keep
# in the columns argument you put the columns whose values you want to make into columns

employee1=employee.pivot_table(index=['Department'],values = ['Salary'], aggfunc = {'Salary':'mean'})
employee1.head()
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.
